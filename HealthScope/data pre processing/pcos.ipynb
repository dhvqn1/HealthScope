```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Undiagnosed PCOS Likelihood', data=df, hue='Menstrual Regularity', palette='coolwarm', alpha=0.6)
plt.title('Age vs. Undiagnosed PCOS Likelihood (Colored by Menstrual Regularity)')
plt.xlabel('Age')
plt.ylabel('Undiagnosed PCOS Likelihood')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.figure(figsize=(8, 5))
sns.countplot(x='Menstrual Regularity', data=df, palette='viridis', hue='Menstrual Regularity', legend=False)
plt.title('Distribution of Menstrual Regularity')
plt.xlabel('Menstrual Regularity')
plt.ylabel('Count')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid") # Set a nice style for the plots

# Create a histogram for the 'Age' column
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
import pandas as pd
df = pd.read_csv('pcos_data.csv')

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")

plt.figure(figsize=(8, 6))
sns.boxplot(x='Family History of PCOS', y='Lifestyle Score', data=df, palette='pastel')
plt.title('Lifestyle Score Distribution by Family History of PCOS')
plt.xlabel('Family History of PCOS')
plt.ylabel('Lifestyle Score')
plt.show()
```
